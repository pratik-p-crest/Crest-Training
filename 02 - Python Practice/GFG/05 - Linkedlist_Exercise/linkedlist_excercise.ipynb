**Ex - 1: Search an Element in a Circular Linked List**

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return

        
        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        temp.next = new_node
        new_node.next = self.head

    def search(self, key):
        if not self.head:
            return False

        temp = self.head
        while True:
            if temp.data == key:
                return True
            temp = temp.next
            if temp == self.head:
                break

        return False
```

```python
cll = CircularLinkedList()
cll.append(10)
cll.append(20)
cll.append(30)

print(cll.search(20))  
print(cll.search(40))  
```

**Ex - 2: Pretty print Linked List**

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def pretty_print(self):
        if not self.head:
            print("Empty Linked List")
            return

        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

```

```python
ll = LinkedList()
ll.append(10)
ll.append(20)
ll.append(30)

ll.pretty_print()
```

**Ex - 3: Reverse a linked list using Stack**

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def reverse_using_stack(self):
        if not self.head or not self.head.next:
            return

        stack = []
        temp = self.head

        while temp:
            stack.append(temp)
            temp = temp.next

        self.head = stack.pop()
        temp = self.head

        while stack:
            temp.next = stack.pop()
            temp = temp.next

        temp.next = None

    def pretty_print(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
```

```python
ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
ll.append(4)

print("Before Reverse:")
ll.pretty_print()

ll.reverse_using_stack()

print("After Reverse:")
ll.pretty_print()
```

**Ex - 4: Stack using Doubly Linked List**

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


class Stack:
    def __init__(self):
        self.top = None

    def push(self, data):
        new_node = Node(data)

        if self.top:
            new_node.next = self.top
            self.top.prev = new_node

        self.top = new_node

    def pop(self):
        if not self.top:
            print("Stack Underflow")
            return None

        popped = self.top
        self.top = self.top.next

        if self.top:
            self.top.prev = None

        return popped.data

    def peek(self):
        if not self.top:
            print("Stack is Empty")
            return None
        return self.top.data

    def display(self):
        temp = self.top
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")


# Example usage
s = Stack()
s.push(10)
s.push(20)
s.push(30)

s.display()

print("Popped:", s.pop())
print("Top:", s.peek())

s.display()
```
