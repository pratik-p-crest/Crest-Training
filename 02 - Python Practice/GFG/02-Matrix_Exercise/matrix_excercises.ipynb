**Ex - 1 : Python program to a Sort Matrix by index-value equality count**

Given a Matrix, the task is to write a Python program that can sort its rows or columns on a measure of the number of values equal to its index number. For each row or column, count occurrences of equality of index number with value. After computation of this count for each row or column, sort the matrix on basis of the count extracted for each row or column in step 1.

```python
matrix = [
  [0, 1, 2],
  [1, 1, 1,],
  [0, 2, 2]
]

def get_index_value_equality_count(matrix):
    
    index_value_equality_count = {}

    count_of_row = 0
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if j == matrix[i][j]:
                count_of_row += 1
        index_value_equality_count[i] = count_of_row
        count_for_row = 0

    sorted_index_value_equality_count = dict(sorted(index_value_equality_count.items(), key=lambda item: item[1]))

    return sorted_index_value_equality_count


def sort_matrix_rows(matrix,counts):
    sorted_matrix = []

    for j in counts.keys():
        sorted_matrix.append(matrix[j])

    return sorted_matrix

counts = get_index_value_equality_count(matrix)
sorted_matrix = sort_matrix_rows(matrix,counts)

print('Before sort :', matrix)
print('\nAfter sort :', sorted_matrix)
```

**Ex - 3: Python Program to Convert String Matrix Representation to Matrix**

```python
def string_to_matrix(s):
    s = s.strip()[2:-2] 
    rows = s.split("], [")

    matrix = []
    for row in rows:
        matrix.append(list(map(int, row.split(", "))))

    return matrix

s = "[[1, 2, 3], [4, 5, 6], [7, 8, 9]]"
print(string_to_matrix(s))
```

**Ex - 9: Adding and Subtracting Matrices in Python**

```python
def add_matrices(A, B):
    result = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] + B[i][j])
        result.append(row)
    return result


def subtract_matrices(A, B):
    result = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] - B[i][j])
        result.append(row)
    return result


A = [ [1, 2, 3], [4, 5, 6]]

B = [[7, 8, 9], [1, 2, 3]]

print("Addition:", add_matrices(A, B))
print("Subtraction:", subtract_matrices(A, B))
```
