**1. Random Value Generation**

Write a Python program to generate a random color hex, a random alphabetical string, random value between two integers (inclusive) and a random multiple of 7 between 0 and 70. Use random.randint()

```python
import random
import string

random_hex = "#{:06x}".format(random.randint(0, 0xFFFFFF))

random_string = ''.join(random.choice(string.ascii_letters) for _ in range(8))

random_int = random.randint(10, 50)

random_multiple_7 = random.randint(0, 10) * 7

print("Random Hex Color:", random_hex)
print("Random Alphabetical String:", random_string)
print("Random Integer (10 to 50):", random_int)
print("Random Multiple of 7 (0 to 70):", random_multiple_7)
```

**2. Floor and Ceiling Rounding**

Write a Python program to configure the rounding to round to the floor, ceiling. Use decimal.ROUND_FLOOR, decimal.ROUND_CEILING


```python
from decimal import Decimal, ROUND_FLOOR, ROUND_CEILING

num = Decimal("12.67")

floor_value = num.quantize(Decimal("1"), rounding=ROUND_FLOOR)

ceiling_value = num.quantize(Decimal("1"), rounding=ROUND_CEILING)

print("Original Number :", num)
print("Floor Rounding  :", floor_value)
print("Ceiling Rounding:", ceiling_value)
```
