**1. Move Key to End**

Write a Python program to create an OrderedDict with the following key-value pairs:

'Laptop': 40

'Desktop': 45

'Mobile': 35

'Charger': 25

Now move the 'Desktop' key to the end of the dictionary and print the updated contents.

```python
from collections import OrderedDict
```

```python
items = OrderedDict([
    ('Laptop', 40),
    ('Desktop', 45),
    ('Mobile', 35),
    ('Charger', 25)
])

print("Original OrderedDict:")
print(items)

items.move_to_end('Desktop')

print("\nUpdated OrderedDict (after moving 'Desktop' to end):")
print(items)
```
