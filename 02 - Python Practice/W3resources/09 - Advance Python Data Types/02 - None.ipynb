**5. Uppercase Conversion with None Check**

Write a Python program that takes a user's input and converts it to uppercase. If the input is empty, return None

```python
def uppercase():

    str = input('Enter string:')

    if len(str) == 0:
        return None
    else:
        str = str.upper()
        return str
```

```python
uppercase()
```

```python
print(uppercase())
```
