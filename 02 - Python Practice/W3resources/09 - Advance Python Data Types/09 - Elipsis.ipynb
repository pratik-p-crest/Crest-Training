1.Write a Python function process_list(lst) that:

If no list is passed, use ... as a default value

If the argument is ..., print "No list provided"

Otherwise, return the sum of the list elements

```python
def process_list(lst=...):
    if lst is ...:
        print("No list provided")
    else:
        return sum(lst)


# Test cases
print(process_list())     
print(process_list([1, 2, 3, 4]))
```
