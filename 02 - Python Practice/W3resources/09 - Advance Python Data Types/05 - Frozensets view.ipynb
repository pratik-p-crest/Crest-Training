**1. Frozenset and Set Conversion**

Write a Python program that converts a "frozen set" to a regular set and vice versa. Compare the differences between these two data types.

```python

normal_set = {1, 2, 3, 4}

frozen = frozenset(normal_set)

print("Frozenset:", frozen)

converted_set = set(frozen)

print("Converted back to set:", converted_set)
```

frozen sets are immutable
