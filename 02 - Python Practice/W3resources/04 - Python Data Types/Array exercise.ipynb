**1. Create an Array and Access Elements**

Write a Python program to create an array of 5 integers and display the array items. Access individual elements through indexes.

```python
import array

a = array.array('i',[1,2,3,4,5,6])
print(a,'\n')

for i in a:
    print(i)
```
