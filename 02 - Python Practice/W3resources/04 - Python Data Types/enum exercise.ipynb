**3. Display Enum member names ordered by their values**

Write a Python program to display all the member names of an enum class ordered by their values.

```python
class country(Enum):
    Afghanistan = 93
    Albania = 355
    Algeria = 213
    Andorra = 376
    Angola = 244
    Antarctica = 672

ordered_members = sorted(country, key=lambda member: member.value)

for i in ordered_members:
    print(i.name)
```
