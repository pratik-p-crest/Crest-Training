**1. Write a Python program to get the latest number of confirmed, deaths, recovered and active cases of Novel Coronavirus (COVID-19) Country wise.**

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-17-2020.csv"
covid_data = pd.read_csv(url)

print("First 5 rows:\n")
print(covid_data.head())

print("\nDataset information:\n")
print(covid_data.info())

print("\nMissing values in each column:\n")
print(covid_data.isna().sum())
```
