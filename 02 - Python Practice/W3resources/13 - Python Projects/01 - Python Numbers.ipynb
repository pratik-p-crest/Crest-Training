```python
FIRST_NAMES = [
    "Aarav", "Vivaan", "Aditya", "Rohan", "Karan",
    "Priya", "Ananya", "Neha", "Pooja", "Kavya"
]

LAST_NAMES = [
    "Sharma", "Verma", "Patel", "Singh", "Mehta",
    "Gupta", "Iyer", "Reddy", "Chopra", "Malhotra"
]

COUNTRIES = [
    "India", "USA", "Canada", "UK", "Australia"
]

DOMAINS = [
    "gmail.com", "yahoo.com", "outlook.com"
]
```

```python
import random

def generate_first_name():
    return random.choice(FIRST_NAMES)

def generate_last_name():
    return random.choice(LAST_NAMES)

def generate_full_name():
    return f"{generate_first_name()} {generate_last_name()}"

def generate_username(first, last):
    return f"{first.lower()}{last.lower()}{random.randint(10,99)}"

def generate_email(username):
    return f"{username}@{random.choice(DOMAINS)}"

def generate_profile():
    first = generate_first_name()
    last = generate_last_name()
    username = generate_username(first, last)

    return {
        "first_name": first,
        "last_name": last,
        "full_name": f"{first} {last}",
        "username": username,
        "email": generate_email(username),
        "age": random.randint(18, 60),
        "country": random.choice(COUNTRIES)
    }
```

```python
profile = generate_profile()

print("Random Profile Generated\n")
for key, value in profile.items():
    print(f"{key.capitalize():12}: {value}")
```
