**1. Multiple Threads Creation**

Write a Python program to create multiple threads and print their names.

```python
import threading
```

```python
def print_thread_name():
    print(f"Thread name: {threading.current_thread().name}")

threads = []

for i in range(5):
    t = threading.Thread(target=print_thread_name, name=f"Thread-{i}")
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

**2. Even and Odd Printer Threads**

Write a Python program that creates two threads to find and print even and odd numbers from 30 to 50.

```python
def print_even():
    for num in range(30, 51):
        if num % 2 == 0:
            print(f"Even: {num}")

def print_odd():
    for num in range(30, 51):
        if num % 2 != 0:
            print(f"Odd : {num}")

t1 = threading.Thread(target=print_even, name="EvenThread")
t2 = threading.Thread(target=print_odd, name="OddThread")

t1.start()
t2.start()

t1.join()
t2.join()

print("Done")
```
