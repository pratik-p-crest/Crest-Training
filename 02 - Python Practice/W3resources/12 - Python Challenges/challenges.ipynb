**1. Write a Python program to check if a number is a perfect square.**

Input : 9

Output : True

```python
def is_perfect_square(n):
    if n < 0:
        return False
    r = int(n ** 0.5)
    return r * r == n

print(is_perfect_square(9))   
print(is_perfect_square(10))  
```

**2. Write a Python program to find a missing number from a list.**

Input : [1,2,3,4,6,7,8]

Output : 5

```python
nums = [1, 2, 3, 4, 6, 7, 8]

full_set = set(range(1, max(nums) + 1))
missing_number = full_set - set(nums)

print(missing_number.pop())
```
