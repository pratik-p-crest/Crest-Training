```python
from geopy.geocoders import Nominatim
```

```python
def search_location(place):
    
    geolocator = Nominatim(user_agent="geo_search_app")

    location = geolocator.geocode(place, addressdetails=True)

    if location:
        print("Place Name :", location.address)
        print("Latitude   :", location.latitude)
        print("Longitude  :", location.longitude)

        address = location.raw.get('address', {})
        print("\n--- Address Details ---")
        print("House Number :", address.get('house_number'))
        print("Road         :", address.get('road'))
        print("City         :", address.get('city') or address.get('town'))
        print("State        :", address.get('state'))
        print("Country     :", address.get('country'))
        print("Postal Code :", address.get('postcode'))
    else:
        print("Location not found")

search_location("India Gate, New Delhi")
```
