**Unit Test for Equality of Two Lists**

Write a Python unit test program that checks if two lists are equal.

```python
def check_equality(l1, l2):

    if len(l1) != len(l2):
        
        return "both list are not equal"
    else:
        flag = None
        for i in range(len(l1)):
            if (l1[i] == l2[i]):
                flag = 0
            else:
                flag = 1 
        
        if flag == 0:
            return "list are equal"
        else:
            return "both list are not equal"
```

```python
list1 = [10,20,30,40,50]
list2 = [10,20,30,40,50]

print(check_equality(list1,list2))
```

```python
list1 = [10,20,30,40,50]
list2 = [10,20,30,40,50,60]

print(check_equality(list1,list2))
```
