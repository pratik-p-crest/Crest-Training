```python
try:
    f = open('test.txt','w')
    f.write('first line')
except TypeError:
    print('there was type error')
except OSError:
    print('you have an OS error')
finally:
    print(' i always run')
```

```python
def ask_for_int():
    while True:
        try:
            result = int(input('enter number:'))
        except:
            print('that was not a number')
            continue
        else:
            print('thank you')
            break
        finally:
            print('end of try/except/finally')
```
