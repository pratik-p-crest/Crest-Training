```python
def create_cubs(n):
    result = []
    for i in range(1,n):
        result.append(i**3)
    return result
```

```python
create_cubs(10)
```

```python
for i in create_cubs(10):
    print(i)
```

```python
def gen_fibon(n):
    a= 1
    b= 2

    for i in range(n):
        yield a
        a,b = b, a+b
```

```python
for number in gen_fibon(10):
    print(number)
```

```python
def gen_fibon(n):
    a= 1
    b= 2

    output = []

    for i in range(n):
        output.append(a)
        a,b = b, a+b
    
    return output
```
