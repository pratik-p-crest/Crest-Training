**Timing your code**

```python
def fun1(n):
    return [str(i) for i in range(n)]
```

```python
fun1(10)
```

```python
def fun2(n):
    return list(map(str,range(n)))
```

```python
fun2(10)
```

```python
import time
```

```python
start_time = time.time()

result = fun1(10000000)

end_time = time.time()

elapsed_time = end_time - start_time

print(elapsed_time)
```

```python
start_time = time.time()

result = fun2(10000000)

end_time = time.time()

elapsed_time = end_time - start_time

print(elapsed_time)
```

```python
import timeit
```

```python
stat = '''
fun1(100)
'''
```

```python
setup = '''
def fun1(n):
    return [str(i) for i in range(n)]
'''
```

```python
timeit.timeit(stat,setup,number = 1000000)
```

```python
stat2 = '''
fun2(100)
'''
```

```python
setup2 = '''
def fun2(n):
    return list(map(str,range(n)))
'''
```

```python
timeit.timeit(stat2,setup2,number = 1000000)
```

```python
%%timeit
fun1(1000)
```

```python
%%timeit
fun2(1000)
```

**Zipping unzipping**

```python
f = open('file1.txt','w+')
f.write('one file')
f.close()
```

```python
f = open('file2.txt','w+')
f.write('two file')
f.close()
```

```python
import zipfile
```

```python
comp_file = zipfile.ZipFile('comp_file.zip','w')
```

```python
comp_file.write('file1.txt',compress_type=zipfile.ZIP_DEFLATED)
```

```python
comp_file.write('file2.txt',compress_type=zipfile.ZIP_DEFLATED)
```

```python
comp_file.close()
```

```python
zip_obj = zipfile.ZipFile('comp_file.zip','r')
```

```python
zip_obj.extractall('extracted_content')
```
