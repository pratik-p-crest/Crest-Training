```python
class Circle():

    pi= 3.14
    def __init__(self, radius=1):
        
        self.radius = radius
        self.area = radius * radius * self.pi 
        # pi (class object attribute) can be call as Circle.pi within the class

    def get_circumfrance(self):
        return self.radius * self.pi * 2
```

```python
my_circle = Circle()
```

```python
my_circle.pi
```

```python
my_circle.radius
```

```python
my_circle = Circle(5)
```

```python
my_circle.radius
```

```python
my_circle.get_circumfrance()
```

```python
my_circle.area
```

**Polymorphism**

```python
class Dog():

    def __init__(self, name):
        self.name = name

    def speake(self):
        return self.name + ' say woof!'
```

```python
class cat():

    def __init__(self, name):
        self.name = name

    def speake(self):
        return self.name + ' say meow!'
```

```python
niko = Dog('niko')
felix = cat('fliex')
```

```python
niko.speake()
```

```python
felix.speake()
```

```python
for pet in [niko, felix]:
    print(pet)
    print(type(pet))
```

```python
pet_speak(niko)
```

```python
class Animal():

    def __init__(self, name):
        self.name  =  name

    def speak(self):
        raise NotImplementedError('Subclass must implement this abstract method')
```

```python
myanimal = Animal('fred')
```

```python
fido = Dog('fido')
```

```python
class Book():

    def __init__(self, title, author, pages):

        self.title = title
        self.author = author
        self.pages = pages
    
    def __str__(self):
            return f"{self.title} by {self.author}"
    
    def __len__(self):
         return self.pages
    
    def __del__(self):
         print('book object has been deleted')
```

```python
b = Book('python rocks', 'pratik',200)
```

```python
print(b)
```

```python
str(b)
```

```python
len(b)
```
